```python
#Our MVP focuses on predictions of games on APril 2 - This selects those games.
game_date = (df['GAME_DATE'] == '2023-04-02')
game_date_filtered_df = df[game_date]
game_date_filtered_df = game_date_filtered_df.reset_index(drop=True)
game_date_filtered_df
```

```python
#For each team we collect whwat their last 10 games were prior to April 2 for the players who we are predicting as part of MVP

game_id_filter = (df['TEAM_ID'] == 1610612764)
filtered_df = df[game_id_filter]
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```

```python
new_df = filtered_df.drop_duplicates(subset=['TEAM_ID','GAME_ID','GAME_DATE'])
new_df = new_df.loc[:, ['TEAM_ID','GAME_ID','GAME_DATE']]
new_df = new_df.reset_index(drop=True)
new_df
```

```python
last_10_games = new_df.iloc[3:13:]
last_10_games
```

```python
game_ids = last_10_games['GAME_ID'].tolist()
filtered_df = df.loc[(df['TEAM_ID'] == 1610612764) & (df['GAME_ID'].isin(game_ids))]

# Group the filtered dataframe by player and calculate the average of PTS, REB, STL, and AST
team_grouped_df = filtered_df.groupby(['TEAM_ID','PLAYER_ID', 'PLAYER_NAME']).mean()[['PTS', 'REB', 'STL', 'AST','FGA','FG_PCT','FG3A','FTA',]].reset_index()

# Print the grouped dataframe

#team_grouped_df = team_grouped_df[1:]
#team_grouped_df = team_grouped_df[team_grouped_df['PLAYER_NAME'] != 'Kyrie Irving']
team_grouped_df = team_grouped_df.dropna()
team_grouped_df = team_grouped_df[team_grouped_df['PTS'] != 0]
team_grouped_df.to_csv('WAS_grouped_df_last_10games.csv', index=False)
team_grouped_df
```

```python
#THis code is used for testing to determine the MSEs and actual vs. predicted points for players who played in the APril 2 games.
#This code was used in R file for output analysis.


filtered_df = df[df['GAME_DATE'] >= '2023-01-01']
filtered_df = filtered_df.iloc[:,0:8]
games_by_season['GAME_ID'] = games_by_season['GAME_ID'].astype(int)
filtered_df = pd.merge(filtered_df, games_by_season[['GAME_ID', 'MATCHUP']], on='GAME_ID', how='inner')
filtered_df['WHERE_PLAYED'] = filtered_df['MATCHUP'].apply(lambda x: 'AWAY_GAME' if '@' in x else 'HOME_GAME')
filtered_df[['team1', 'team2']] = filtered_df['MATCHUP'].str.split('@|vs.', expand=True)

filtered_df['team1'] = filtered_df['team1'].str.strip()
filtered_df['team2'] = filtered_df['team2'].str.strip()

filtered_df.loc[filtered_df['TEAM_ABBREVIATION'] == filtered_df['team1'], 'ABBREVIATION'] = filtered_df['team2']
filtered_df.loc[filtered_df['TEAM_ABBREVIATION'] != filtered_df['team1'], 'ABBREVIATION'] = filtered_df['team1']
```
